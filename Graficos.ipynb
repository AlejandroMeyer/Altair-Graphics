### GRÁFICO DE BARRAS SENCILLO
Hacer un gráfico con la población de un país en especifico (México).

```python
datos_pais = data[ data['country'] == "Mexico"]
datos_pais.head(5)
```

```python
# "alt" de altair, necesito un gráfica (chart) del data set (datos_pais)
# donde la forma del gráfico (mark_barr) va a ser de barras y
# vas a codificar (encode) mis datos en esta forma:
# donde "X2 serán los años y "Y" será la población.
```

```python
alt.Chart(datos_pais).mark_bar().encode(   
    x = 'year',
    y = 'pop',
)                                            
```

### GRÁFICO DE BARRAS +

```python
# 1.- "O" de ordinales para que nos de valores de mnera ordenada, categorias.
# 2.- Quiero que uses este valor como "Q" de cuantitativos aunque seán números.
# 1 y 2.- Agrego el objeto "X" y "Y" para agregar más infromación.
# Notas.- Puedo personalizar más mi gráfico, REVISAR DOCUMENTACIÓN.
# 2.- Puedo dar diferente formato a mis ejes con "axis" (checar docuemntación para ver todos los formatos.)
# 2.- El formato ".2s" es para decirle que muestre los valores en millones.
# 3.- Agrego propiedades para cambiar el gráfico en sí.
# 4.- Añado título de mi gráfico.
# 5.- El ancho de mí gráfico (el valor siempre estará en píxeles).
# 1.- Con el objeto axis "labelAngle" modifico la orientación de mis etiquetas.
# Nota.- Las funciones van con minusculas y los objetos en mayusculas.
# 5.- Le decimos a altair, una vez hecho el gráfico y con sus propiedades, concatena que el titulo tenga caracteristicas.
# 0.- Altair puede reconocer algunos nombre de colores o con el #RGB.
# 2.- "grid" sirve para quitar o poner las líneas de fondo del gráfico
```

```python
alt.Chart(datos_pais).mark_bar(color = "#227888").encode(  #0  
    x = alt.X('year:O', title="", axis = alt.Axis(labelAngle= 0, labelFontSize=10)), # 1      
    y = alt.Y('pop:Q', title="Población", axis = alt.Axis(format=".2s", grid=False)), # 2       
).properties(  # 3
    title = "Población de México de 1955 al 2005", # 4
    width = 400,  # 5
).configure_title( #6
    fontSize = 18,
)                              
                                                
```

# GRÁFICOS AVANZADOS

```python
# Puedo sobre escribir gráficos. Algunas veces resulta más fácil crear un gráfico y despues extenderlo.
# 1.- Creo mi gráfico.
# 2.- Le asigno una variable.
# 3.- Extiendo mi gráfico (como si hiciera otro gráfico, ya que son como capas y se puende encimar).
```

```python
# 1 y 2
barras = alt.Chart(datos_pais).mark_bar(color = "#227888").encode(  
    x = alt.X('year:O', title="", axis = alt.Axis(labelAngle= 0, labelFontSize=10)),      
    y = alt.Y('pop:Q', title="", axis = alt.Axis(format=".2s", grid=False, labels = False,  domain=False, ticks=False)),       
)
```

```python
barras
```

```python
etiquetas = barras.mark_text(dy = -5).encode( # 3  el 'dy' es para separar el texto (ver documentación)
    text = alt.Text("pop:Q", format=".2s"),
)
```

```python
etiquetas
```

```python
grafico_final = barras + etiquetas
```

```python
grafico_final.properties(  
    title = "Población de México de 1955 al 2005", 
    width = 400,  
).configure_title( 
    fontSize = 18,
).configure_view(
    strokeWidth=0   # Quita el recuadro de nuestro gráfico
)
   # al fin personalizo los dos gráficos juntos con properties directamente a la variable donde estan juntos los gráficos.
```

### Facetas

```python
# Gráfico de la población total de nuestro data
# Una forma de crear gráficos uno al lado de otro, cuando estas usando 
# los mismos datos. Una faceta es escencialmente diferentes vistas dentro del mimso dataset
```

```python
mis_paises = ['Dominican Republic', 'Mexico', 'Peru']
```

```python
datos_de_mis_paises = data[data['country'].isin(mis_paises)]  # Otra forma de filtrar es con 'isin' esto es de pandas
```

```python
datos_de_mis_paises
```

```python
alt.Chart(datos_de_mis_paises).mark_bar(color = "#227888").encode(    
    x = alt.X('year:O', title="", axis = alt.Axis(labelAngle= 0, labelFontSize=10)), # 1      
    y = alt.Y('pop:Q', title="", axis = alt.Axis(format=".2s", grid=False)),
    column = alt.Column('country', title = "Población de 1955 al 2005")      
).properties(  
    width = 400,  
).configure_title( 
    fontSize = 18,
).configure_view(
    strokeWidth=0   
)                                                   
```

### Facetas Tooltip

```python
# 1.- Un tooltip es un texto que sale en los gráficos, los hace interactivos.
# 2.- Para poner más tooltip se pasa una lista
```

```python
alt.Chart(datos_de_mis_paises).mark_bar(color = "#227888").encode(    
    x = alt.X('year:O', title="", axis = alt.Axis(labelAngle= 0, labelFontSize=10)),     
    y = alt.Y('pop:Q', title="", axis = alt.Axis(format=".2s", grid=False)),
    column = alt.Column('country', title = "Población de 1955 al 2005"),
    tooltip = [alt.Tooltip('pop:Q', format=',.5s', title='Población'), alt.Tooltip('year:O', title='Año')]     # 1 y 2 
).properties(  
    width = 400,  
).configure_title( 
    fontSize = 18,
).configure_view(
    strokeWidth=0   
)     
```

### Interactividad

```python
# Ejemplo de como usar el filtro de selección 'intervalo'
```

```python
from vega_datasets import data
```

```python
source = data.cars()  #Cargamos data set de cars

brush = alt.selection(type='interval')  # Manera en la que comunica que ha sido seleccionado dentro de unos de esos graficos.
```

```python
# Vrush es pincel
# 1.- Gráfico de dispersión
# 2.- Creamos una condición en los colores, lo que no esta dentro de la seleccion le vamos a dar color lightgray
# alt.value es para indicalre a altair que le estamos asignando un color y no valor, que no lo busque como columna
# Le damos un color basado en la columna "Orgin"

# 3.- Creamos gráfico de barras
# 4.- Color basado en origen
# 5.- Cuantas filas hay en origen (como cantidad)
# 6.- Si los datos estan dentro de la selección aplicas lo anterior
# 7.- '&' Concantena verticalmente y '|' concatena horizontal
```

```python
points = alt.Chart(source).mark_point().encode(  # 1
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
    color=alt.condition(brush, 'Origin:N', alt.value('lightgray')) # 2
).add_selection(
    brush
)
# Son gráficos separados
bars = alt.Chart(source).mark_bar().encode(  # 3 
    y='Origin:N',
    color='Origin:N', # 4
    x='count(Origin):Q'  # 5
).transform_filter(  # 6
    brush
)

points & bars # 7
```

### Gráficp de ejemplos: Barras con valores negativos, Mapas, Matriz, etc.

```python
# Datum es el singular de datos
# 1.- Altair, quiero que veas cada uno de los datos de esta línea del dataframe, quiero que
# tomes el atributo 'nonfarm_change (que es la columna) y sí es mayor que 0
# vas a ponerle el color azul, si no cumple la condición entonces será naranja
```

```python
from vega_datasets import data
```

```python
source = data.us_employment()

alt.Chart(source).mark_bar().encode(
    x="month:T",
    y="nonfarm_change:Q",
    color=alt.condition(
        alt.datum.nonfarm_change > 0, #1
        alt.value("steelblue"),  # The positive color
        alt.value("orange")  # The negative color
    )
).properties(width=600)
```
